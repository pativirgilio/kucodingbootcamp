```python
import os
import csv
import pandas as pd
```

```python
bike1 = "201801_citibikejc_tripdata.csv"
bike2 = "201802_citibikejc_tripdata.csv"
bike3 = "201803_citibikejc_tripdata.csv"
bike4 = "JC-201804-citibike-tripdata.csv"
bike5 = "JC-201805-citibike-tripdata.csv"
bike6 = "JC-201806-citibike-tripdata.csv"
```

```python
bike1_df = pd.read_csv(bike1, encoding="ISO-8859-1")
bike2_df = pd.read_csv(bike2, encoding="ISO-8859-1")
bike3_df = pd.read_csv(bike3, encoding="ISO-8859-1")
bike4_df = pd.read_csv(bike4, encoding="ISO-8859-1")
bike5_df = pd.read_csv(bike5, encoding="ISO-8859-1")
bike6_df = pd.read_csv(bike6, encoding="ISO-8859-1")
```
